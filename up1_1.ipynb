**Задача**: парадокс монти-холла, посчитать вероятность (в процентах целых чисел), что человек выиграл бы, если бы сменил решение
```
def monty_hall(correct_door_number, participant_guesses):
    pass
```
Пример:
```
monty_hall(1, [3,3,2,3,3,2,2,3,2,2,1,1,1,1]) == 71
```

````python
def monty_hall(correct_door_number, participant_guesses):
    count = 0
    for guess in participant_guesses:
        if guess != correct_door_number:
            count += 1
    return (count * 100) // len(participant_guesses)

monty_hall(1, [3, 3, 2, 3, 3, 2, 2, 3, 2, 2, 1, 1, 1, 1])
````

**Задача**: известно, что несколько различных кодонов (триплетов, состоящий из трёх оснований) ДНК могут кодировать одну и ту же аминокислоту. Каждый белок закодирован последовательностью аминокислот. Подробнее:

[DNA_codon_table](https://en.wikipedia.org/wiki/DNA_codon_table)

[Генетический код](https://ru.wikipedia.org/wiki/%D0%93%D0%B5%D0%BD%D0%B5%D1%82%D0%B8%D1%87%D0%B5%D1%81%D0%BA%D0%B8%D0%B9_%D0%BA%D0%BE%D0%B4) – (но тут таблица для РНК-кодонов)

Мы хотим написать функцию, которая проверяла бы, кодирует ли 2 последовательности ДНК один и тот же белок. Считаем, что в функцию приходят корректные строки (число элементов кратно 3, строки одинаковой длины и т.п.)

```
def code_for_same_protein(seq1, seq2):
    pass
```

Пример:
```
code_for_same_protein("ATGTCGTCAATTTAA","ATGTCGTCAATTTAA") is True
code_for_same_protein("ATGTTTTAA","ATGTTCTAA") is True
code_for_same_protein("ATGTTTTAA","ATGATATAA") is False
code_for_same_protein("ATGTTTTAA","ATGATATAA") is False
code_for_same_protein("ATGTTTGGGAATAATTAAGGGTAA","ATGTTCGGGAATAATGGGAGGTAA") is False
```

````python
codons = {
    'AAA': 'K', 'TAT': 'Y', 'ATG': 'M', 'GAG': 'E', 
    'TTA': 'L', 'GTC': 'V', 'GCG': 'A', 'ACG': 'T', 
    'GCC': 'A', 'AGT': 'S', 'GAC': 'D', 'GTT': 'V',
    'CCA': 'P', 'ACC': 'T', 'CGC': 'R', 'CGA': 'R',
    'GTG': 'V', 'CTA': 'L', 'TCG': 'S', 'GCT': 'A',
    'CTT': 'L', 'GGG': 'G', 'AAT': 'N', 'CAT': 'H',
    'ATC': 'I', 'ACT': 'T', 'CTG': 'L', 'TAG': '*',
    'TGG': 'W', 'TAC': 'Y', 'GAT': 'D', 'CCT': 'P',
    'TCT': 'S', 'CGT': 'R', 'ATT': 'I', 'TTC': 'F',
    'TCA': 'S', 'CGG': 'R', 'CAA': 'Q', 'ACA': 'T',
    'AAG': 'K', 'TTT': 'F', 'AGA': 'R', 'CCC': 'P',
    'AGC': 'S', 'CTC': 'L', 'GCA': 'A', 'GGA': 'G',
    'TAA': '*', 'ATA': 'I', 'TGA': '*', 'CAC': 'H',
    'TTG': 'L', 'GGT': 'G', 'TGC': 'C', 'CAG': 'Q',
    'GGC': 'G', 'AGG': 'R', 'GTA': 'V', 'CCG': 'P',
    'AAC': 'N', 'TCC': 'S', 'GAA': 'E', 'TGT': 'C'
}
````

````python
def code_for_same_protein(seq1, seq2):
    i = 0
    while i < len(seq1):
        if (codons[seq1[i:(i + 3)]] != codons[seq2[i:(i + 3)]]):
            return False
        i += 3
    return True

print(code_for_same_protein("ATGTCGTCAATTTAA", "ATGTCGTCAATTTAA"))
print(code_for_same_protein("ATGTTTTAA", "ATGTTCTAA"))
print(code_for_same_protein("ATGTTTTAA", "ATGATATAA"))
print(code_for_same_protein("ATGTTTTAA", "ATGATATAA"))
print(code_for_same_protein("ATGTTTGGGAATAATTAAGGGTAA", "ATGTTCGGGAATAATGGGAGGTAA"))
````

````python
import numpy as np
import matplotlib.pyplot as plt
````

$$X = \{\sqrt{a^2 + 4} + \sqrt{b^2 + 4} \mid a + b = 3; a,b \in \mathbb{R}\}$$

````python
def X(a):
    b = 3 - a
    return np.sqrt(a ** 2 + 4) + np.sqrt(b ** 2 + 4)
````

````python
%matplotlib inline
a = np.linspace(-10, 10, 200)
y = X(a)

fig, ax = plt.subplots()

line, = ax.plot(a, y, lw=2)

ax.annotate('$\inf X$', xy=(1.5, 5), xytext=(0, 20),
            arrowprops=dict(facecolor='red', shrink=0.05))
plt.show()
````

````python
def X2(a, b):
    return np.sqrt(a ** 2 + 4) + np.sqrt(b ** 2 + 4)
````

````python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.gca(projection='3d')

x = np.linspace(-10, 10, 20)
y = np.linspace(-10, 10, 20)
X, Y = np.meshgrid(x, y)
Z = X2(X, Y)

a = x
b = 3 - a
z = np.repeat(0, 20)
ax.plot(a, b, z)
z2 = X2(a, b)
ax.plot(a, b, z2)


ax.contour3D(X, Y, Z, 50, cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
````
